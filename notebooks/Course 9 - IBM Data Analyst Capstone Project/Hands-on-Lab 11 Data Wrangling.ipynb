#### Step 1: Import the necessary module.


### 1. Load the Dataset


<h5>1.1 Import necessary libraries and load the dataset.</h5>


Ensure the dataset is loaded correctly by displaying the first few rows.


```python
# Import necessary libraries
import pandas as pd
import numpy as np

# Load the Stack Overflow survey data
dataset_url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/n01PQ9pSmiRX6520flujwQ/survey-data.csv"
df = pd.read_csv(dataset_url)

# Display the first few rows
print(df.head())
```

#### 2. Explore the Dataset


<h5>2.1 Summarize the dataset by displaying the column data types, counts, and missing values.</h5>


```python
# Write your code here
#Display Column Data Types
print('Data Types of the Dataframe columns: ')
print(df.dtypes)
print('-------------------------')

#Display the count of Dataframe
print('Number of columns and rows of the Dataframe')
print('Rows: ',df.shape[0], 'Columns: ', df.shape[1])
print('-------------------------')

#Display missing values of the Dataframe
df_missing_values = df.isna().sum()
print('Missing values of a dataframe:')
print(df_missing_values)
```

<h5>2.2 Generate basic statistics for numerical columns.</h5>


```python
# Write your code here
print(df.describe())
```

### 3. Identifying and Removing Inconsistencies


<h5>3.1 Identify inconsistent or irrelevant entries in specific columns (e.g., Country).</h5>


```python
# Write your code here
#Count unique countries
print('=== Unique Country Entries ===')
print(df['Country'].unique())

#Count the frequency of each unique country
print('=== Country Value Counts ===')
print(df['Country'].value_counts(dropna = False))
```

<h5>3.2 Standardize entries in columns like Country or EdLevel by mapping inconsistent values to a consistent format.</h5>


```python
## Write your code here

df['Country'] = df['Country'].str.strip().str.title()
df['EdLevel'] = df['EdLevel'].str.strip().str.title()

# Verify standardization
print("=== Unique Countries ===")
print(df['Country'].unique())

print("\n=== Unique Education Levels ===")
print(df['EdLevel'].unique())
```

### 4. Encoding Categorical Variables


<h5>4.1 Encode the Employment column using one-hot encoding.</h5>


```python
## Write your code here
print("=== Unique Countries ===")
#print(df['Employment'].unique())

# One-hot encode the 'Employment' column
employment_dummies = pd.get_dummies(df['Employment'], prefix='Employment')

# Add the new dummy columns to the original dataframe
df = pd.concat([df, employment_dummies], axis=1)

print("\n=== Sample of One-Hot Encoded Columns ===")
df.head()
```

### 5. Handling Missing Values


<h5>5.1 Identify columns with the highest number of missing values.</h5>


```python
## Write your code here

# Count missing values per column
missing_values = df.isna().sum()
print(missing_values)
# Sort columns by descending number of missing values
missing_counts_sorted = missing_values.sort_values(ascending = False)

# Display columns with the highest missing values
print("=== Columns with Missing Values (sorted) ===")
print(missing_counts_sorted)

# Optional: just show top 5
print("\n=== Top 5 Columns with Most Missing Values ===")
print(missing_counts_sorted.head(5))
```

<h5>5.2 Impute missing values in numerical columns (e.g., `ConvertedCompYearly`) with the mean or median.</h5>


```python
## Write your code here

# Preview missing values in 'ConvertedCompYearly'
print("Missing values before imputation:", df['ConvertedCompYearly'].isnull().sum())

mean_comp = df['ConvertedCompYearly'].mean()
df['ConvertedCompYearly'] = df['ConvertedCompYearly'].fillna(mean_comp)
print('Missing value of ConvertedCompYearly after imputation: ', df['ConvertedCompYearly'].isnull().sum())
```

<h5>5.3 Impute missing values in categorical columns (e.g., `RemoteWork`) with the most frequent value.</h5>


```python
## Write your code here
# Preview missing values in 'RemoteWork'
print("Missing values before imputation:", df['RemoteWork'].isnull().sum())

most_frequent_value = df['RemoteWork'].mode()[0]
df['RemoteWork'] = df['RemoteWork'].fillna(most_frequent_value)
print('Missing value of RemoteWork after imputation: ', df['RemoteWork'].isnull().sum())
```

### 6. Feature Scaling and Transformation


<h5>6.1 Apply Min-Max Scaling to normalize the `ConvertedCompYearly` column.</h5>


```python
## Write your code here


#Normalizing ConvertedCompYearly
min_comp = df['ConvertedCompYearly'].min()
max_comp = df['ConvertedCompYearly'].max()

df['ConvertedCompYearly_MinMax'] = (df['ConvertedCompYearly'] - min_comp) / (max_comp - min_comp)

print(df[['ConvertedCompYearly','ConvertedCompYearly_MinMax']].head())
```

<h5>6.2 Log-transform the ConvertedCompYearly column to reduce skewness.</h5>


```python
## Write your code here

df['ConvertedCompYearly_Log'] = np.log1p(df['ConvertedCompYearly'])
print(df[['ConvertedCompYearly', 'ConvertedCompYearly_MinMax', 'ConvertedCompYearly_Log']].head(5))
```

### 7. Feature Engineering


<h5>7.1 Create a new column `ExperienceLevel` based on the `YearsCodePro` column:</h5>


```python
## Write your code here
# Handle missing or non-numeric values

most_frequent_value = df['YearsCodePro'].mode()[0]
df['YearsCodePro'] = df['YearsCodePro'].fillna(most_frequent_value)
print('Missing value of YearsCodePro after imputation: ', df['YearsCodePro'].isnull().sum())

df['YearsCodePro'] = pd.to_numeric(df['YearsCodePro'], errors='coerce')

# Create ExperienceLevel column
def experience_level(years):
    if years <= 2:
        return 'Junior'
    elif years <= 5:
        return 'Intermediate'
    elif years <= 10:
        return 'Senior'
    else:
        return 'Expert'

df['ExperienceLevel'] = df['YearsCodePro'].apply(experience_level)

# Check the result
print(df[['YearsCodePro', 'ExperienceLevel']].head(10))
```

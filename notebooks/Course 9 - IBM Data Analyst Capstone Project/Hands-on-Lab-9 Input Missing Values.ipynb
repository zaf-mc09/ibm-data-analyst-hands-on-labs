### Step 1: Import Required Libraries


```python
import pandas as pd
```

### Step 2: Load the Dataset Into a Dataframe


#### **Read Data**
<p>
The functions below will download the dataset into your browser:
</p>


```python
file_path = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/n01PQ9pSmiRX6520flujwQ/survey-data.csv"
df = pd.read_csv(file_path)

# Display the first few rows to ensure it loaded correctly
print(df.head())
```

### Step 3. Finding and Removing Duplicates
##### Task 1: Identify duplicate rows in the dataset.


```python
## Write your code here

duplicates = df[df.duplicated(keep = False)]
num_duplicates = duplicates.shape[0]

print('Number of duplicate rows: ')
print(duplicates.head())
```

##### Task 2: Remove the duplicate rows from the dataframe.



```python
## Write your code here

drop_duplicate = df.drop_duplicates()

df_remaining_duplicate = drop_duplicate.duplicated().sum

print ('Number of duplicate rows after removal: ')
print(df_remaining_duplicate)
```

### Step 4: Finding Missing Values
##### Task 3: Find the missing values for all columns.


```python
## Write your code here
missing_value = df.isna().sum()
print('Missing values per column')
print(missing_value)
```

##### Task 4: Find out how many rows are missing in the column RemoteWork.


```python
## Write your code here
missing_value_remote = df['RemoteWork'].isna().sum()
print('Missing values of remote work')
print(missing_value_remote)
```

### Step 5. Imputing Missing Values
##### Task 5: Find the value counts for the column RemoteWork.


```python
## Write your code here

df_value_counts = df['RemoteWork'].value_counts()
print('Value counts of RemoteWork: ')
print(df_value_counts)
```

##### Task 6: Identify the most frequent (majority) value in the RemoteWork column.



```python
## Write your code here
most_frequent_remote = df['RemoteWork'].mode()[0]
print('Most frequent value in RemoteWork')
print(most_frequent_remote)
```

##### Task 7: Impute (replace) all the empty rows in the column RemoteWork with the majority value.



```python
## Write your code here
df['RemoteWork'] = df['RemoteWork'].fillna(most_frequent_remote)
print('Missing value after imputating RemoteWork')
print(df['RemoteWork'].isna().sum())
```

```python
## Write your code here
print('Distribution of ConvertedCompYearly: ')
print(df['ConvertedCompYearly'].describe())
```
